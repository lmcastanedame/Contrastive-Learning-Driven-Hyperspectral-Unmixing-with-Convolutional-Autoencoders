## Method load_HSI
Loads the HSI and reference endmembers

```python
def load_HSI(path):
    try:
        data = sio.loadmat(path)
    except NotImplementedError:
        data = hdf.File(path, 'r')

    Y = np.asarray(data['Y'], dtype=np.float32)
    GT = np.asarray(data['GT'], dtype=np.float32)
    if Y.shape[0] < Y.shape[1]:
        Y = Y.transpose()
    Y = Y / np.max(Y.flatten())
    n_bands = Y.shape[1]
    n_rows = data['lines'].item()
    n_cols = data['cols'].item()
    Y = np.reshape(Y, (n_cols, n_rows, n_bands))
    return Y, GT
```

## Method training_input_fn
Extracts patches for training 

```python
# def training_input_fn(hsi, patch_size, patch_number, batch_size):
#     patches = extract_patches_2d(hsi, (patch_size, patch_size), max_patches=patch_number)
#     return patches

def training_input_fn_tf(hsi, patch_size, patch_number):
    # Convert the image to a TensorFlow tensor if it's not already
    hsi = tf.convert_to_tensor(hsi, dtype=tf.float32)
    
    # Extract patches for the image (here we use a single channel to make it compatible)
    patches = tf.image.extract_patches(
        images=tf.expand_dims(hsi, axis=0),  # Add batch dimension
        sizes=[1, patch_size, patch_size, 1],
        strides=[1, patch_size, patch_size, 1],  # Non-overlapping patches
        rates=[1, 1, 1, 1],
        padding="VALID"
    )
    
    # Reshape to the desired patch_number
    patches = tf.reshape(patches, [-1, patch_size, patch_size, hsi.shape[-1]])[:patch_number]
    
    return patches.numpy()

# def training_input_fn(hsi, patch_size, patch_number, batch_size):
#     patches = extract_patches_2d(hsi, (patch_size, patch_size), max_patches=patch_number)
#     # Generate a single random alpha and repeat it for each patch
#     alpha = 1 # np.random.uniform(0.8, 1.2)  # Single alpha value for all patches
#     alphas = alpha * np.ones(patches.shape[0])  # Create an array with the same length as patches
#     return patches, alphas
```

## Class SumToOne
Enforces the ASC. Regularizations on the abundance maps go in here. 

```python
class SumToOne(tf.keras.layers.Layer):
    def __init__(self, params, **kwargs):
        super(SumToOne, self).__init__(**kwargs)
        self.num_outputs = params['num_endmembers']
        self.params = params
    
    def build(self, input_shape):
        assert len(input_shape) >= 2
        input_dim = input_shape[-1]
        
    def call(self, x):
        x = tf.nn.softmax(self.params['scale'] * x)
        return x
```

```python
# class Encoder(tf.keras.Model):
#     def __init__(self, params):
#         super(Encoder, self).__init__()
#         self.params = params
#         self.hidden_layer_one = tf.keras.layers.Conv2D(
#             filters=self.params['e_filters'],
#             kernel_size=self.params['e_size'],
#             activation=self.params['activation'], strides=1, padding='same',
#             kernel_initializer=params['initializer'], use_bias=False)
#         self.hidden_layer_two = tf.keras.layers.Conv2D(
#             filters=self.params['num_endmembers'], 
#             kernel_size=1,
#             activation=self.params['activation'], 
#             strides=1, 
#             padding='same',
#             kernel_initializer=self.params['initializer'], 
#             use_bias=False)
#         self.asc_layer = SumToOne(params=self.params, name='abundances')

#     def call(self, input_patch):
#         code = self.hidden_layer_one(input_patch)
#         code = tf.keras.layers.BatchNormalization()(code)
#         code = tf.keras.layers.SpatialDropout2D(0.2)(code)
#         code = self.hidden_layer_two(code)
#         code = tf.keras.layers.BatchNormalization()(code)
#         code = tf.keras.layers.SpatialDropout2D(0.2)(code)
#         code = self.asc_layer(code)
#         return code


class Encoder(tf.keras.Model):
    def __init__(self, params):
        super(Encoder, self).__init__()
        self.params = params
        self.hidden_layer_one = tf.keras.layers.Conv2D(
            filters=self.params['e_filters'],
            kernel_size=self.params['e_size'],
            activation=self.params['activation'],
            strides=1,
            padding='same',
            kernel_initializer=params['initializer'],
            use_bias=False
        )
        self.batch_norm1 = tf.keras.layers.BatchNormalization()
        self.dropout1 = tf.keras.layers.SpatialDropout2D(0.2)

        self.hidden_layer_two = tf.keras.layers.Conv2D(
            filters=self.params['num_endmembers'],
            kernel_size=1,
            activation=self.params['activation'],
            strides=1,
            padding='same',
            kernel_initializer=self.params['initializer'],
            use_bias=False
        )
        self.batch_norm2 = tf.keras.layers.BatchNormalization()
        self.dropout2 = tf.keras.layers.SpatialDropout2D(0.2)
        
        self.asc_layer = SumToOne(params=self.params, name='abundances')

    def call(self, input_patch, training=False):
        code = self.hidden_layer_one(input_patch)
        code = self.batch_norm1(code, training=training)
        code = self.dropout1(code, training=training)
        code = self.hidden_layer_two(code)
        code = self.batch_norm2(code, training=training)
        code = self.dropout2(code, training=training)
        code = self.asc_layer(code)
        return code
```

## Class Decoder
Extends the Layer class. Decodes abundance patches and reconstructs the inputs to the encoder.

```python
class Decoder(tf.keras.layers.Layer):
    def __init__(self, params):
        super(Decoder, self).__init__()
        self.output_layer = tf.keras.layers.Conv2D(filters=params['d_filters'], kernel_size=params['d_size'],
                                                   activation='linear',
                                                   kernel_constraint=tf.keras.constraints.non_neg(),
                                                   name='endmembers', strides=1, padding='same',
                                                   kernel_regularizer=None,
                                                   kernel_initializer=params['initializer'], use_bias=False)

    def call(self, code):
        recon = self.output_layer(code)
        return recon

    def getEndmembers(self):
        return self.output_layer.get_weights()
```

```python
class Autoencoder(tf.keras.Model):
    def __init__(self, params):
        super(Autoencoder, self).__init__()
        self.encoder = Encoder(params)
        self.decoder = Decoder(params)
        self.params = params

    def call(self, patch):
        abunds = self.encoder(patch)
        # tf.summary.histogram('abunds', abunds, step=1)
        #         abunds = tf.keras.layers.SpatialDropout2D(0.08)(abunds)
        output = self.decoder(abunds)
        return output

    def getEndmembers(self):
        endmembers = self.decoder.getEndmembers()[0]
        if endmembers.shape[1] > 1:
            endmembers = np.squeeze(endmembers).mean(axis=0).mean(axis=0)
        else:
            endmembers = np.squeeze(endmembers)
        return endmembers

    def getAbundances(self, hsi):
        return np.squeeze(self.encoder.predict(np.expand_dims(hsi, 0)))

    def train(self, patches, callback):
        self.plotWhileTraining = callback
        self.fit(patches, patches, epochs=self.params['epochs'], batch_size=self.params['batch_size'],
                 callbacks=[self.plotWhileTraining], verbose=0)

    def saveResults(self, fname):
        endmembers = self.getEndmembers()
        abundances = self.getAbundances(params['data'])
        sads = self.plotWhileTraining.sads
        sio.savemat(fname, {'M': endmembers, 'A': abundances, 'sads': sads})
```

## SAD and Loss functions

```python
def SAD(y_true, y_pred):
    y_true = tf.math.l2_normalize(y_true, axis=-1)
    y_pred = tf.math.l2_normalize(y_pred, axis=-1)
    A = (y_true * y_pred)
    sad = tf.math.acos(A)
    return sad

def numpy_SAD(y_true, y_pred):
    return np.arccos(y_pred.dot(y_true) / (np.linalg.norm(y_true) * np.linalg.norm(y_pred)))

def loss(model, original):
    reconstruction_error = SAD(model(original), original)
    return reconstruction_error
```

## Various functions for training and plotting

```python
def train(loss, model, opt, original):
    with tf.GradientTape() as tape:
        gradients = tape.gradient(loss(model, original) + sum(model.losses), model.trainable_variables)
        gradient_variables = zip(gradients, model.trainable_variables)
        opt.apply_gradients(gradient_variables)


def order_endmembers(endmembers, endmembersGT):
    num_endmembers = endmembers.shape[0]
    dict = {}
    sad_mat = np.ones((num_endmembers, num_endmembers))
    for i in range(num_endmembers):
        endmembers[i, :] = endmembers[i, :] / endmembers[i, :].max()
        endmembersGT[i, :] = endmembersGT[i, :] / endmembersGT[i, :].max()
    for i in range(num_endmembers):
        for j in range(num_endmembers):
            sad_mat[i, j] = numpy_SAD(endmembers[i, :], endmembersGT[j, :])
    rows = 0
    while rows < num_endmembers:
        minimum = sad_mat.min()
        index_arr = np.where(sad_mat == minimum)
        if len(index_arr) < 2:
            break
        index = (index_arr[0][0], index_arr[1][0])
        if index[0] in dict.keys():
            sad_mat[index[0], index[1]] = 100
        elif index[1] in dict.values():
            sad_mat[index[0], index[1]] = 100
        else:
            dict[index[0]] = index[1]
            sad_mat[index[0], index[1]] = 100
            rows += 1
    ASAM = 0
    num = 0
    for i in range(num_endmembers):
        if np.var(endmembersGT[dict[i]]) > 0:
            ASAM = ASAM + numpy_SAD(endmembers[i, :], endmembersGT[dict[i]])
            num += 1

    return dict, ASAM / float(num)


def plotEndmembers(endmembers):
    endmembers = endmembers / endmembers.max()
    fig = plt.figure(1)
    for i in range(num_endmembers):
        ax = plt.subplot(2, 2, i + 1)
        plt.plot(endmembers[i, :], 'r', linewidth=1.0)
        ax.get_xaxis().set_visible(False)
    plt.tight_layout()
    plt.show()


def plotEndmembersAndGT(endmembers, endmembersGT):
    num_endmembers = endmembers.shape[0]
    n = num_endmembers // 2  # how many digits we will display
    if num_endmembers % 2 != 0: n = n + 1
    dict, sad = order_endmembers(endmembers, endmembersGT)

    fig = plt.figure(num=1, figsize=(8, 8))
    plt.clf()
    title = "aSAM score for all endmembers: " + format(sad, '.3f') + " radians"
    st = plt.suptitle(title)
    for i in range(num_endmembers):
        endmembers[i, :] = endmembers[i, :] / endmembers[i, :].max()
        endmembersGT[i, :] = endmembersGT[i, :] / endmembersGT[i, :].max()

    for i in range(num_endmembers):
        ax = plt.subplot(2, n, i + 1)
        plt.plot(endmembers[i, :], 'r', linewidth=1.0)
        plt.plot(endmembersGT[dict[i], :], 'k', linewidth=1.0)
        ax.set_title("SAD: " + str(i) + " :" + format(numpy_SAD(endmembers[i, :], endmembersGT[dict[i], :]), '.4f'))
        ax.get_xaxis().set_visible(False)

    plt.tight_layout()
    st.set_y(0.95)
    fig.subplots_adjust(top=0.88)
    plt.draw()
    plt.pause(0.001)


def plotAbundancesSimple(abundances):
    abundances = np.transpose(abundances, axes=[1, 0, 2])
    num_endmembers = abundances.shape[2]
    n = num_endmembers // 2
    if num_endmembers % 2 != 0: n = n + 1
    fig = plt.figure(2, figsize=[8, 8])
    for i in range(num_endmembers):
        ax = plt.subplot(2, n, i + 1)
        divider = make_axes_locatable(ax)
        cax = divider.append_axes(position='bottom', size='5%', pad=0.05)
        im = ax.imshow(abundances[:, :, i], cmap='viridis')
        plt.colorbar(im, cax=cax, orientation='horizontal')
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        im.set_clim([0, 1])
    plt.tight_layout()
    plt.draw()
    plt.pause(0.001)


class PlotWhileTraining(tf.keras.callbacks.Callback):
    def __init__(self, model, plot_every_n, hsi, gt, patience=10):
        super(PlotWhileTraining, self).__init__()
        self.model = model
        self.plot_every_n = plot_every_n
        self.num_endmembers = gt.shape[0]
        self.input = hsi
        self.endmembersGT = gt
        self.sads = []
        self.losses = []
        self.best_sad = float('inf')
        self.wait = 0
        self.patience = patience
        self.should_stop = False
        self.best_endmembers = None  # To store the best endmembers
        self.best_abundances = None  # To store the best abundances

    def on_train_begin(self, logs={}):
        self.losses = []
        self.val_loss = []
        self.sads = []

    def on_epoch_end(self, epoch, logs=None):
        current_loss = logs.get('SAD')
        self.losses.append(current_loss)
        print('*', end='')
    
        endmembers = self.model.getEndmembers()
        endmembers = np.squeeze(endmembers)
        if self.endmembersGT is not None:
            _, sad = order_endmembers(endmembers, self.endmembersGT)
            self.sads.append(sad)
    
            # Early stopping logic
            if sad < self.best_sad:
                self.best_sad = sad
                self.wait = 0  # Reset wait if there's improvement
                # Save the best endmembers and abundances
                self.best_endmembers = endmembers
                self.best_abundances = self.model.getAbundances(self.input)
            else:
                self.wait += 1
                if self.wait >= self.patience:
                    print(f"\nEarly stopping at epoch {epoch+1}")
                    self.should_stop = True  # Set the stopping flag
                    self.model.stop_training = True  # Stop the training loop
    
            if self.plot_every_n != 0 and epoch % self.plot_every_n == 0:
                plotEndmembersAndGT(self.endmembersGT, endmembers)
                abundances = self.model.getAbundances(self.input)
                plotAbundancesSimple(abundances)

    def get_best_results(self):
        """Retrieve the best endmembers and abundances after training."""
        return self.best_endmembers, self.best_abundances
```

## Data set (Augmentations)

```python
# Function for random cropping and resizing
def random_crop_resize(image, prob):
    if prob == 1:
        h, w = image.shape[:2]  # Get height and width of the image
        
        # Calculate 95% crop size
        crop_h = int(0.95 * h)
        crop_w = int(0.95 * w)
        
        # Randomly select top-left corner for cropping
        top = random.randint(0, h - crop_h)
        left = random.randint(0, w - crop_w)
        
        # Crop the image
        cropped_image = image[top:top + crop_h, left:left + crop_w, :]
        
        # Resize it back to the original size
        image = tf.image.resize(cropped_image, [h, w])
    
    return image

# Function for random horizontal, vertical, or both flips
def random_flip(image, prob):
    if prob == 1:
        # Randomly select flip type: 0 = horizontal, 1 = vertical, 2 = both
        flip_type = random.choice([0, 1, 2])
        
        if flip_type == 0:
            # Horizontal flip
            image = tf.image.flip_left_right(image)
        elif flip_type == 1:
            # Vertical flip
            image = tf.image.flip_up_down(image)
        elif flip_type == 2:
            # Both flips
            image = tf.image.flip_left_right(image)
            image = tf.image.flip_up_down(image)
    return image

# Function for color jitter (brightness and contrast)
def random_brightness_contrast(image, prob, alpha_range=(0.8, 1.2), beta_range=(-0.2, 0.2)):
    if prob == 1:
        # Sample alpha (contrast) from the given range
        alpha = np.random.uniform(alpha_range[0], alpha_range[1])
        
        # Sample beta (brightness) from the given range
        beta = np.random.uniform(beta_range[0], beta_range[1])
        
        # Adjust the image pixel values based on the formula I_{3}^c(i,j) = alpha * I_{2}^c(i,j) + beta
        image = alpha * image
    return image

kernel_cache = {}

def gaussian_kernel(size: int, sigma: float):
    """Generates a 2D Gaussian kernel."""
    x = tf.range(-size // 2 + 1, size // 2 + 1, dtype=tf.float32)
    y = tf.range(-size // 2 + 1, size // 2 + 1, dtype=tf.float32)
    xx, yy = tf.meshgrid(x, y)
    kernel = tf.exp(-(xx**2 + yy**2) / (2.0 * sigma**2))
    kernel = kernel / tf.reduce_sum(kernel)
    return kernel[:, :, tf.newaxis, tf.newaxis]

def get_gaussian_kernel(size, sigma):
    """Retrieves a cached Gaussian kernel or generates a new one if not cached."""
    key = (size, sigma)
    if key not in kernel_cache:
        kernel_cache[key] = gaussian_kernel(size, sigma)
    return kernel_cache[key]

def random_gaussian_blur(image, prob, min_kernel_size=3, max_kernel_size=7, min_sigma=0.1, max_sigma=2.0):
    if prob == 1:
        # Choose a random odd kernel size within the specified range
        kernel_size = random.choice(range(min_kernel_size, max_kernel_size + 1, 2))
        
        # Choose a random standard deviation within the specified range
        sigma = random.uniform(min_sigma, max_sigma)
        
        # Retrieve or create a Gaussian kernel
        gaussian_kernel_tensor = get_gaussian_kernel(kernel_size, sigma)
        
        # Split channels and apply the Gaussian blur to each channel individually
        channels = tf.split(image, num_or_size_splits=image.shape[-1], axis=-1)
        blurred_channels = [
            tf.nn.conv2d(channel[tf.newaxis, ...], gaussian_kernel_tensor, strides=[1, 1, 1, 1], padding='SAME')
            for channel in channels
        ]
        
        # Concatenate the blurred channels back together
        image = tf.concat([tf.squeeze(channel, axis=0) for channel in blurred_channels], axis=-1)
        
    return image

# Main augmentation function to apply all transformations
def apply_augmentation(image, p1=0, p2=0, p3=0, p4=0):
    
    # Random crop and resize
    image = random_crop_resize(image, prob=p1)
    
    # Horizontal flip
    image = random_flip(image, prob=p2)
    
    # Color jitter (brightness and contrast)
    image = random_brightness_contrast(image, prob=p3)
    
    # Gaussian blur
    image = random_gaussian_blur(image, prob=p4)
    
    return image
```

```python
def augment_image(transform, image):
    with tf.compat.v1.Session() as session:
        augmented_image = apply_augmentation(
            image, 
            p1=transform["crop_prob"], 
            p2=transform["flip_prob"], 
            p3=transform["brightness_prob"], 
            p4=transform["blur_prob"]
        )
        return augmented_image.eval(session=session)
```

## Spectral Augmentations

```python
from scipy.ndimage import gaussian_filter1d

# Spectral jitter (adds random noise to the spectral bands)
def spectral_jitter(image, prob, noise_factor=0.01):
    if prob == 1:
        noise = np.random.randn(*image.shape) * noise_factor
        image = image + noise
    return image

# Spectral shift (shifts spectral bands)
def spectral_shift(image, prob, shift_range=2):
    if prob == 1:
        shift = np.random.randint(-shift_range, shift_range)
        image = np.roll(image, shift, axis=-1)  # Shifting along the spectral axis
    return image

# Band dropping (drops random spectral bands)
def band_dropping(image, prob, drop_probability=0.1, original_depth=162):
    if prob == 1:
        mask = np.random.rand(image.shape[-1]) > drop_probability
        dropped_image = image[:, :, mask]
        
        # Pad to original depth if needed
        if dropped_image.shape[-1] < original_depth:
            pad_width = original_depth - dropped_image.shape[-1]
            dropped_image = np.pad(dropped_image, ((0, 0), (0, 0), (0, pad_width)), mode='constant')
        
        return dropped_image
    return image

# Spectral smoothing (applies a Gaussian filter to smooth the spectral dimension)
def spectral_smoothing(image, prob, sigma=1):
    if prob == 1:
        image = gaussian_filter1d(image, sigma=sigma, axis=-1)
    return image

# Main spectral augmentation function
def apply_spectral_augmentation(image, p1=0, p2=0, p3=0, p4=0):
    # Spectral jitter
    image = spectral_jitter(image, prob=p1)
    
    # Spectral shift
    image = spectral_shift(image, prob=p2)
    
    # Band dropping
    image = band_dropping(image, prob=p3)
    
    # Spectral smoothing
    image = spectral_smoothing(image, prob=p4)
    
    return image
```

```python
# Function to calculate SAD between two endmember vectors
def calculate_sad(endmember1, endmember2):
    endmember1 = endmember1 / np.linalg.norm(endmember1)
    endmember2 = endmember2 / np.linalg.norm(endmember2)
    sad_value = np.arccos(np.clip(np.dot(endmember1, endmember2), -1.0, 1.0))
    return sad_value

# Function to find the optimal ordering of predicted endmembers using SAD
def match_abundances(predicted, true):
    num = true.shape[2]
    cost_matrix = np.zeros((num, num))
    for i in range(num):
        for j in range(num):
            cost_matrix[i, j] = mean_squared_error(predicted[:, :, i], true[:, :, j])
    # print(cost_matrix)
    row_ind, col_ind = linear_sum_assignment(cost_matrix)
    reordered_predicted = predicted[col_ind]
    return reordered_predicted, row_ind, col_ind
```
